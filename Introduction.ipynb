```python
#Building a program for 10Alytics

#String
print("Welcome to 10Alytics")
name = input("What is your first name: ") #by default input function gives a string
education = input("What did you study in college/university: ")

#Integers (wholenumbers(can be positive or negative))

Year_of_birth = int(input("Please put in your year of birth: "))#since input function gives a string, you need to convert it to an integer so python sees it as a number hence the ([int] function in the code)
age = 2026 - Year_of_birth
print(age)

#Float - rate your skillset

python = float(input("What is your proficiency in Python on a scale of (0 -10): ")) #better to use a float when you answer can either be an integer or a float
excel = float(input("What is your proficiency in Excel on a scale of (0 -10): "))
tableau = float(input("What is your proficiency in Tableau on a scale of (0 -10): "))
sql = float(input("What is your proficiency in SQL on a scale of (0 -10): "))

#List
course = input("Provide the list of courses you are interested in (separate each course with a comma(,)): ") 
course_list = course.split(", ") # to convert from a string to a list

#Set
skills = input ("Enter your skillset separated by a comma (e.g Python, Machine Learning, Excel, Airflow, etc.): ")
skill_list = skills.split(", ")
skill_set = set(skill_list) #set will keep it unique

#Dictionary {key:value}
full_name = input("What is your full name?: ")
location = input("What is your current location?: ")
phone_num = input("Please enter your phone number: ")
email = input("What is your email address: ")

full_details = {"Full_name" : full_name, "Location" : location, "Phone Number" : phone_num, "Email address" : email }

#Boolean
full_time = input("Do you want to learn full time? (yes/no): ")
answer = full_time.lower()
full_time_bool = answer =="yes"

full_time_bool

print("----------------------------------------------------------------------------------")
print(f"Welcome to 10Alytics, {name}, you studied {education} at the university.") 
print(f"Your year of birth is {Year_of_birth} which makes you {age} years old.")
print(f"Python:{python}, Excel:{excel}, Tableau: {tableau}, SQL: {sql}")
print(f"Your course preferences {course_list}")
print (f"Your skill set are: {skill_set}")
print(f"Your full details are: {full_details}")
print(f"full time learning: {full_time_bool}")
```

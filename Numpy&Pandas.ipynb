## 1.2 Indexing and Slicing

```python
#Indexing is that we want a particular item from our dataset (simple values)
#slicing means we want a group of items...(multiple values)

temperatures = np.array([36.6,37.1,36.9,38.2,37.0])
print("First Temperature:", temperatures[0])
print("Last Temperature:", temperatures [-1])
print("First three readings:", temperatures[0:3])  # this can also be done using [:3])
print("Last three readings:", temperatures [-3:])

#Boolean indexing
fever_threshold = 37.5
fevers = temperatures [temperatures > fever_threshold]
print("fevers_readings(>37.5)", fevers)
```

```python
# Column labels: [Maths, English, Science]
scores_list = [[85, 90, 78], 
               [92, 88, 95], 
               [70, 75, 80], 
               [88, 82, 91]]
scores = np.array(scores_list)

print("Student 1's Scores:", scores[0]) # or
scores_list[0]
```

## 1.3 Vectorized Operation

```python
# creating vectors (vectors are important in numpy because it makes it easier to carry out bulk operations)

import time
import numpy as np  # Added missing numpy import

big_list = list(range(1_000_000))
big_array = np.arange(0, 1_000_001)

#in Python loop
start = time.time()
squared_list = [x**2 for x in big_list]
python_time = time.time() - start

#Numpy vectorized
start = time.time()
squared_array = big_array**2
numpy_time = time.time() - start

print(f"Python Time: {python_time:.4f} seconds")
print(f"Numpy Time: {numpy_time:.4f} seconds")
python_time/numpy_time
```

## 1.4 Aggregation functions
##### axis = 0 Operate down the rows(collapse the rows, keep the columns)(top to bottom)
##### axis = 1 Operate down the columns and keep the rows)(left to right)

```python
# With this function, you can count data, sum data, etc
print("Score per subject", np.sum(scores, axis = 0))
print ("Score per person", np.sum (scores, axis = 1))

print("Average score per subject", np.average(scores, axis = 0))
print("Average score per person", np.average(scores, axis = 1))

print ("Minimum score per subject", np. min(scores, axis = 0))
print ("Minimum score per person", np. min(scores, axis = 1))
print ("Maximum score per subject", np. max(scores, axis = 0))
print ("Maximum score per person", np. max(scores, axis = 1))
       
```
